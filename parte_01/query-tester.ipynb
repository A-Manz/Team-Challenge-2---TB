```python
import pandas as pd
import sqlite3
```

```python
# Connectar con la db
conn = sqlite3.connect('db/sql-murder-mystery.db')
cursor = conn.cursor()

def sql_query(query):
    cursor.execute(query)
    ans = cursor.fetchall()
    names = [description[0] for description in cursor.description]
    return pd.DataFrame(ans,columns=names)
```
